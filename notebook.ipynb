```python
import pandas as pd

df = pd.read_csv("edges.csv", sep=',', usecols=[2])

df['name'] = df['name'].apply(lambda st: [s.strip() for s in st.split('(interacts with)')])
df = df['name'].apply(pd.Series)


df_nodes = pd.read_csv("nodes.csv", sep=',', usecols=[1, 2])

display(df)
display(df_nodes)
```

```python
import networkx as nx

G = nx.Graph()

for row, series in df_nodes.iterrows():
    G.add_node(series[1], n_rowerow=series[0])

for row, series in df.iterrows():
    G.add_edge(series[0], series[1])

nx.draw(G)
```
